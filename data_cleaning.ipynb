```python
#One-hot encoding
category_list=['REV_SETT_1','REV_SETT_2','REV_SETT_3','REV_SETT_4','REV_SETT_5','REV_SETT_6',
               'RESTRUCTURED_PLAN','ST_GENERAL_ACC','GENDER','PRODUCT','NATURE_OF_BUS_DESC']

df_one_hot_encoded = pd.get_dummies(df, columns=category_list).astype("float")
```

```python
df_one_hot_encoded.to_csv("final_df.csv", index=False)
```
